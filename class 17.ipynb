# Positional Argument

```python
def add(num1 ,num2 , num3):
    print(f'num1: {num1}, num2 : {num2}, num3:  {num3}')
    return num1 + num2 + num3
```

```python
add(1 ,3 ,4)
```

# key word argument

```python
#def add(num1 ,num2 , num3):

add( num3=3, num1=4, num2=5)
```

Default parameter values

```python
def add(num1=1 ,num2=2 , num3=3):
    print(f'num1: {num1}, num2 : {num2}, num3:  {num3}')
    return num1 + num2 + num3
```

```python
add(4 ,7 ,8 )
```
