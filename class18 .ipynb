```python
def make_album(artist, album, song=None):
    if song:
        result = {artist: album, 'song':song}
    else:
        result = {artist: album}
    return result 
```

```python
def enough (cap, on, wait):
    x = cap - on
    if x > wait:
       return  0
    else :
        return (on + wait - cap)
```

```python
enough(10 , 5 ,5)
```

```python
enough(100, 60, 50)
```
